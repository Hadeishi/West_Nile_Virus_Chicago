## Instead of making this refer to the previous column, refer to the first colum displaced n days

```python
def time_machine(df, column, days, bc = 0):
    new_df = df[['Date', column]]
    for i in range(1,days+1):
        new_col_name = column + "_" + str(i) +"_back"
        new_col = one_day_back(df, column, i, bc)
        new_df[new_col_name] = new_col
    return new_df

def one_day_back(df, column, day, bc):
    col = df[column].tolist()
    for i in range(day):
        del col[-1]
        col.insert(0, bc)
    return col
```

```python
time_machine(nminus_df, 'relative_humidity', 5)
```
